```python
epgm = Cartographer()

#white brackground, easier to make out map  
epgm.Background(25,25,"OOO","white")

#meltwater
epgm.UpdateArea(10,0,15,5,"OOO","blue")
epgm.UpdateArea(11,5,13,9,"OOO","blue")
epgm.UpdateArea(13,9,14,12,"OOO","blue")
epgm.UpdateArea(13,12,13,15,"OOO","blue")

#trees, sigle tile
for x,y in zip([2,3,8,16,24,5,9,3,2,13,20,23,24,20,14,15,23,19,7,8,17,20,20,4,8],
               [22,4,9,17,20,14,9,20,17,23,4,3,10,4,8,9,2,20,22,13,1,9,14,8,1]):
    epgm.UpdateArea(x-1,y-1,x+1,y+1,"OOO","green")
    epgm.UpdateSquare(x,y,"BBB","brown")
    
#cave 
epgm.UpdateArea(10,16,14,18,"BBB","gray")
epgm.UpdateSquare(13,16,"OOO","black")

for x,y in zip([8,15,15,12,11,10,17,11],
               [19,22,12,15,15,19,13,12]):
    epgm.UpdateSquare(x,y,"BBB","gray")
    
    
#misc rocks
epgm.UpdateArea(12,3,13,4,"BBB","gray")

#river
epgm.UpdateSquare(13,7,"OOO","blue")
epgm.UpdateSquare(13,8,"OOO","blue")
epgm.UpdateSquare(13,9,"OOO","blue")
```

# 3.0 Svaroh Temple of Volhynian - Altair

```python
stov = Cartographer()
stov.Background(10,10,"OOO",l="tan")
stov.UpdateArea(1,2,8,8,b="OOO",l="gold",edge_maker=True)

#front doors
stov.UpdateSquare(3,1,b="BBB",l="white")
stov.UpdateSquare(4,1,b="PNS",l="brown")
stov.UpdateSquare(5,1,b="PNS",l="brown")
stov.UpdateSquare(6,1,b="BBB",l="white")
stov.UpdateSquare(4,2,b="OOO",l="gold")
stov.UpdateSquare(5,2,b="OOO",l="gold")

#backrooms
stov.UpdateArea(1,7,8,8,b="OOO",l="gold",edge_maker=True)
stov.UpdateSquare(1,7,b="BW",l="gold")

#pillars + status
for node in [(3,1),(6,1),(2,3),(7,3),(2,6),(7,6)] + [(4,4),(4,5),(5,5),(5,4)]:
    stov.UpdateSquare(node[0],node[1],b="BBB",l="white")
    
#odds and ends 
for node in [(2,8),(3,8),(5,7),(6,7),(7,7)]:
    stov.UpdateSquare(node[0],node[1],b="BBB",l="m")    

```
