## 2.1 Crystal Origins - Wizard Tower

```python
cowt = Cartographer()
cowt.Background(19,4,"OOO","g")
cowt.UpdateArea(2,1,3,1,"BBB",l="black")
cowt.UpdateSquare(2,1,"PWE",l="m")

#sidewinder pattern
for xy in [(3,1),(5,1),(7,1),(9,1),(11,1)]:
    cowt.UpdateSquare(xy[0],xy[1],"CSE",l="white")
for xy in [(3,2),(5,2),(7,2),(9,2),(11,2)]:
    cowt.UpdateSquare(xy[0],xy[1],"CNW",l="white")
for xy in [(4,2),(6,2),(8,2),(10,2),(12,2)]:
    cowt.UpdateSquare(xy[0],xy[1],"CNE",l="white")
for xy in [(4,1),(6,1),(8,1),(10,1),(12,1)]:
    cowt.UpdateSquare(xy[0],xy[1],"CSW",l="white")

cowt.UpdateArea(13,0,16,3,"OOO",l="y")    
cowt.UpdateSquare(13,1,"OOO",l="m") 


cowt.UpdateSquare(13,0,"BW",l="y") 
cowt.UpdateSquare(13,3,"BW",l="y") 

cowt.UpdateSquare(16,0,"BE",l="y") 
cowt.UpdateSquare(16,1,"BE",l="y") 
cowt.UpdateSquare(16,2,"BE",l="y") 
cowt.UpdateSquare(16,3,"BE",l="y") 
```

```python
cowt_mm = SquareGraph(SquareList=cowt.map_holder,CharacterList=[])
cowt_mm.MakeMap()
```

```python
cowt_mm.ShowMap(SaveFig=True,PlotGraph=True,SaveName="WizardTower.png")
!convert WizardTower.png -trim WizardTower.png
```

# 3.0 Svaroh Temple of Volhynian - Altair

```python
stov = Cartographer()
stov.Background(10,10,"OOO",l="tan")
stov.UpdateArea(1,2,8,8,b="OOO",l="gold",edge_maker=True)

#front doors
stov.UpdateSquare(3,1,b="BBB",l="white")
stov.UpdateSquare(4,1,b="PNS",l="brown")
stov.UpdateSquare(5,1,b="PNS",l="brown")
stov.UpdateSquare(6,1,b="BBB",l="white")
stov.UpdateSquare(4,2,b="OOO",l="gold")
stov.UpdateSquare(5,2,b="OOO",l="gold")

#backrooms
stov.UpdateArea(1,7,8,8,b="OOO",l="gold",edge_maker=True)
stov.UpdateSquare(1,7,b="BW",l="gold")

#pillars + status
for node in [(3,1),(6,1),(2,3),(7,3),(2,6),(7,6)] + [(4,4),(4,5),(5,5),(5,4)]:
    stov.UpdateSquare(node[0],node[1],b="BBB",l="white")
    
#odds and ends 
for node in [(2,8),(3,8),(5,7),(6,7),(7,7)]:
    stov.UpdateSquare(node[0],node[1],b="BBB",l="m")    

```

```python
stov_mm = SquareGraph(SquareList=stov.map_holder,CharacterList=[])
stov_mm.MakeMap()
```

```python
stov_mm.ShowMap(SaveFig=False,PlotGraph=True,SaveName="VolhynianSvarohTemple.png")
```
