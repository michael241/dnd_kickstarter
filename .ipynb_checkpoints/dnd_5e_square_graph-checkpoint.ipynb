# 1.0 Initialize Map and Character 

```python
#initalize character
A = Character(name="Alpha"
              ,icon="A"
              ,team=0
              ,con=10
             ,x=0
             ,y=0
             ,speed=5
             ,level=1
             ,hit_dice=Dice(amount=1,side=12)
             )
A.Setup()

B = Character(name="Beta"
              ,icon="B"
              ,team=1
              ,con=8
             ,x=3
             ,y=0
             ,speed=5
             ,level=1
             ,hit_dice=Dice(amount=1,side=8)
             )#.Setup()
B.Setup()


#list characters 
CL = [A,B]

SL  = [
 Square(0,3,b="OOO"),Square(1,3,b="OOO"),Square(2,3,b="OOO"),Square(3,3,b="OOO"),
 Square(0,2,b="OOO"),Square(1,2,b="OOO"),Square(2,2,b="BBB"),Square(3,2,b="OOO"),
 Square(0,1,b="OOO"),Square(1,1,b="BS") ,Square(2,1,b="OOO"),Square(3,1,b="BW"),
 Square(0,0,b="OOO"),Square(1,0,b="CNE"),Square(2,0,b="BBB"),Square(3,0,b="BW")     
]

#make map (with graph connections)
SQ = SquareGraph(SquareList=SL,CharacterList=CL)
SQ.MakeMap()
```
